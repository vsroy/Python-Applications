```python
#Displaying volcanoes in world map
volcano = pd.read_csv('volcano.csv')
```

```python
lat_vo = list(volcano['Latitude'])
lon_vo = list(volcano['Longitude'])
name_vo = list(volcano['Name'])
```

```python
vol = fo.FeatureGroup(name='My Map')
```

```python
for lat,long,name in zip(lat_vo,lon_vo,name_vo):
    vol.add_child(fo.Marker(location=[lat,long],popup=name,icon=fo.Icon(color='red')))
```

```python
map.add_child(vol)
```

```python
#plotting us cities and population in each city
popu = pd.read_csv('us cities pop.csv')
```

```python
popu.head()
```

```python
lat_po = list(popu['lat'])
lon_po = list(popu['lon'])
name_po = list(popu['name'])
pop_po = list(popu['pop'])
```

```python
po = fo.FeatureGroup(name='My Map')
```

```python
def mar(popu):
    if(popu > 35000):
        return 'red'
    elif(popu>10000 and popu <=35000):
        return 'blue'
    else:
        return 'green'
```

```python
for lat,long,name,pop in zip(lat_po,lon_po,name_po,pop_po):
    po.add_child(fo.Marker(location=[lat,long],
                           popup=[pop,name],
                           icon=fo.Icon(color=mar(pop))))
```

```python
map.add_child(po)
```
